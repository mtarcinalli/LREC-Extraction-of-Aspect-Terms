# Resultados

```python
import xmltodict
import pandas as pd
from IPython.core.display import display, HTML
import glob
from collections import Counter
```

```python
i=0
tuples = []
results = {
    'Precision': [], 
    'Recall': [], 
    'F-measure': [],
    'Explicits': [],
    'Implicits': [],    
}
cats = {
    'Event' : [],
    'Feature' : [],
    'Qualification' : [],
    'Contextual' : []
}
subs = {
    'Event:Non-verbal_form' : [],
    'Event:Verb' : [],
    'Feature:Attribute' : [],
    'Feature:Equivalence' : [],
    'Feature:Is-a' : [],
    'Feature:Part-of' : [],
    'Qualification:Adjective' : [], 
    'Qualification:Equivalence' : [],
    'Qualification:Nominal_form' : [],
    'Contextual:Location' : [],
    'Contextual:Related' : [],
}

for arq in glob.glob("*xml"):
    #print(arq)
    f = open(arq,"r")
    xml = f.read()
    res = xmltodict.parse(xml, process_namespaces=True)
    algoritmo = "_".join(arq.split('_')[:-1])
    dominio = arq.split('_')[-1][:-4]
    tuples.append((dominio,algoritmo))
    results['Precision'].append(round(float(res['resultados']['precisao']),2))
    results['Recall'].append(round(float(res['resultados']['revocacao']),2))
    results['F-measure'].append(round(float(res['resultados']['medidaf']),2))
    results['Explicits'].append(round(float(res['resultados']['pExplicitos']),2))
    results['Implicits'].append(round(float(res['resultados']['pImplicitos']),2))
    for cat in cats: cats[cat].append(0)
    for sub in subs: subs[sub].append(0)    
    if float(res['resultados']['pImplicitos']) > 0:
        for cat in res['resultados']['tipos']:
            for sub in res['resultados']['tipos'][cat]:
                #print(cat,sub,res['resultados']['tipos'][cat][sub])
                cats[cat][i] += int(res['resultados']['tipos'][cat][sub])
                subs[cat+':'+sub][i] += int(res['resultados']['tipos'][cat][sub])    
    i+=1
```

## Resultados

```python
index = pd.MultiIndex.from_tuples(tuples, names=["Domain", "Algorithm"])
df = pd.DataFrame(results, index=index)
df = df.sort_index()
df
```
